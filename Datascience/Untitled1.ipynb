```python
import urllib.request
import random
import threading
import string

def bombard(threadId):
    for i in range(1,1000):
        rand =  ''.join(random.choices(string.ascii_uppercase + string.digits, k=4))
        print(str(threadId) +rand)
        url = "https://api.karasayfa.com/api/X/SearchX?page=1&per_page=10&filter="+rand
        resp =  urllib.request.urlopen(url).read()
        print(resp)

for x in range(1,20):
    threading.Thread(target=bombard,args=(x,)).start()

```
