```python
class Tree:
    def __init__(self, name):
        self.name = name
        self.items=[]
    def addBranch(self,item):
        item = item.strip()
        for i in self.items:
            if i.name == item:
                return i
        nw = Tree(item)
        self.items.append(nw)
        return nw
    
```
