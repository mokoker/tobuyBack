```python
import urllib.request
from bs4 import BeautifulSoup
from pyquery import PyQuery as pq
import json
import psycopg2


```

```python
contents = urllib.request.urlopen("https://www.hepsiburada.com/sitemaps/kategoriler/sitemap_1.xml").read()
y=BeautifulSoup(contents,features="xml")
z=y.find_all("loc")
```

```python
class Tree:
    def __init__(self, name):
        self.name = name
        self.items=[]
    def addBranch(self,item):
        item = item.strip()
        for i in self.items:
            if i.name == item:
                return i
        nw = Tree(item)
        self.items.append(nw)
        return nw
    
```

```python
ana = Tree('Ana Sayfa')
```

```python
def GetPage(page):
    print(page)
    try:
        contents = urllib.request.urlopen(page).read()
        d = pq(contents)
        deactive = d('.deactive-category')
        if deactive.size()>0:
            return
        e = d('ul[class="breadcrumbs"] li ')
        count =0
        last =ana
        for i in e:
            count +=1
            if count == 1:
                continue
            x=pq(i)
            last = last.addBranch(x.text().strip())
    except Exception as e:
        print('hata olustu' + page + str(e))
        


    
```

```python
counter = 0
for y in z:
    counter+=1
    if last > counter:
        continue
    last+=1
    print(counter)
    GetPage(y.text)
```
